```python
from pandas_datareader import data as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create list of holdings and import daily adj. close price from Yahoo Finance from January 1, 2019 to last trading day available
assets =  ['AAPL', 'AMZN', 'ARKK', 'BX', 'CMG', 'COST', 'DIS', 'DOCU', 'DPZ', 'FEZ', 'HD', 'LULU', 'MSFT', 'NTDOY', 'OKTA', 'PTON', 'RH', 'SHOP', 'SPY', 'WMT']
df = pd.DataFrame()

for stock in assets:
    df[stock] = web.DataReader(stock, data_source='yahoo',start='2019-1-1' ,end='2020-10-13')['Adj Close']
```

```python
#calculate daily returns
d_returns = df.pct_change()
```

```python
d_returns.plot(figsize=(16,8))
```

```python
d_returns.describe()
```

```python
#create correlation table
d_returns.corr()
```

```python
#create covariance matrix
cov_matrix_d = d_returns.cov()
cov_matrix_d
```

```python
#annualize the covariances by multiplying by 250
cov_matrix_a = cov_matrix_d * 252
cov_matrix_a
```

```python
#weights calculated in Google Sheets using exported data from StockTrak
weights = np.array([0.0570,0.0928,0.0694,0.0529,0.0530,0.0563,0.0414,0.0296,0.0261,0.0498,0.0212,0.0462,0.0773,0.0601,0.0239,0.0237,0.0177,0.0634,0.0851,0.0531])
weights
```

```python
#calculate portfolio variance
port_variance = np.dot(weights.T, np.dot(cov_matrix_a, weights))
port_variance
```

```python
#convert to percentage
print(str(round(port_variance, 4) * 100) + '%')
```

```python
#standard deviation of the overall portfolio = square root of variance
port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix_a, weights)))
port_volatility
```

```python
#convert to percentage
print(str(round(port_volatility, 4) * 100) + '%')
```
